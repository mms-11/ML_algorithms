```python
import pandas as pd

import pydotplus #pip install pydotplus
from sklearn.tree import export_graphviz
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import numpy as np

```

```python
# @title Ler dataset diabetes
df = pd.read_csv('/content/MyDrive/My Drive/diabetes.csv')

df.head()
```

```python
# @title treinar base de dados
from sklearn.model_selection import train_test_split

X = df.drop(columns=['class'])
Y = df['class']

# Separando os dados em treinamento(75%) e teste(25%)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25)
```

```python
# @title visualizar rvore
def tree_graph_to_png(tree, feature_names, png_file_to_save):
    tree_str = export_graphviz(tree, feature_names=feature_names,
                                     filled=True, out_file=None)
    graph = pydotplus.graph_from_dot_data(tree_str)
    graph.write_png(png_file_to_save)
```

```python
# @title Gerar a rvore
dt = DecisionTreeClassifier(criterion = 'entropy',max_depth = 2)

dt.fit(X_train, Y_train)

# tree.plot_tree(dt)

tree_graph_to_png(dt, feature_names=X_train.columns,
                 png_file_to_save='dt.png')
```

```python
from sklearn.metrics import accuracy_score

Yhat = dt.predict(X_test)

print("Acurcia dos dados de teste  : ", accuracy_score(Y_test,Yhat))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Displaying the confusion matrix

cm = confusion_matrix(Y_test, Yhat, labels=dt.classes_)

disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=dt.classes_)

disp.plot()

plt.show()
```
